```python
import pandas as pd
import datetime as dt
```

```python
file = pd.read_csv('Resources/cities.csv')
```

```python
new_df = pd.DataFrame(file)
```

```python
new_df['City'] = new_df['City'].str.capitalize()
```

```python
new_df['Date'] = pd.to_datetime(new_df.Date, infer_datetime_format=True, format="%Y%m%d")
```

```python
new_df['Date'] = new_df['Date'].dt.date
```

```python
new_df.to_html('cities.html', index=False)
```
